```python
import pandas as pd
```

```python
data = pd.read_csv("/content/Stress.csv")
```

```python
data.head(2)
```

```python
data = data[["Stress Level", "Job Satisfaction","Turnover Intention"]]
data.loc[data["Turnover Intention"] == "Yes", "Turnover Intention"] = 1
data.loc[data["Turnover Intention"] == "No", "Turnover Intention"] = 0
```

```python
data
```

```python
!pip install catboost
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
```

```python
x = data[["Stress Level", "Job Satisfaction"]]
y = data["Turnover Intention"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = CatBoostClassifier(
    iterations=100,      # Количество итераций
    learning_rate=0.1,   # Скорость обучения
    depth=6              # Максимальная глубина деревьев
)
```

```python
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
```

```python
model.predict([1,2])
```

```python
import pickle
with open('model.pkl', 'wb') as f:
  pickle.dump(model, f)
```
